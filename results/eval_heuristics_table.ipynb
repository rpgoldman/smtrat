```python
import pandas as pd
import benchmax.evaluation as ev

def get_data(solvers):
    param = {}
    for solver in solvers:
        param["stats_smtrat_15m_6Gi_"+solver+".xml"] = {"smtrat-static": solver, "smtrat-"+solver: solver}
    return ev.xmls_to_pandas(param,['mcsat-explanation-nlsat_explanation_called','mcsat-explanation-onecellcad_explanation_called',
    'mcsat-explanation-onecellcad_resultant_barriers', 'mcsat-explanation-onecellcad_max_degree', 'mcsat-explanation-onecellcad_num_of_coefficients', 'mcsat-explanation-onecellcad_num_of_discriminants', 'mcsat-explanation-onecellcad_num_of_resultants', 'mcsat-explanation-onecellcad_max_degree_total', 'mcsat-explanation-onecellcad_num_of_coefficients_total', 'mcsat-explanation-onecellcad_num_of_discriminants_total', 'mcsat-explanation-onecellcad_num_of_resultants_total', 'mcsat-explanation-onecellcad_num_of_levels'])

df = get_data(["MCSATOCLWH11","MCSATOCLWH12","MCSATOCLWH13"])

for solver in ["MCSATOCLWH11","MCSATOCLWH12","MCSATOCLWH13"]:
    df.loc[:,(solver,'oc_called')] = df[(solver,'mcsat-explanation-onecellcad_explanation_called')].fillna(0).astype(int)
    df.loc[:,(solver,'nl_called')] = df[(solver,'mcsat-explanation-nlsat_explanation_called')].fillna(0).astype(int)

    df.loc[:,(solver,'resultant_barriers')] = df[(solver,'mcsat-explanation-onecellcad_resultant_barriers')].fillna(0).astype(int)
    
    df.loc[:,(solver,'max_degree')] = df[(solver,'mcsat-explanation-onecellcad_max_degree')].fillna(0).astype(int)
    df.loc[:,(solver,'max_degree_total')] = df[(solver,'mcsat-explanation-onecellcad_max_degree_total')].fillna(0).astype(int)

    df.loc[:,(solver,'num_of_coefficients')] = df[(solver,'mcsat-explanation-onecellcad_num_of_coefficients')].fillna(0).astype(int)
    df.loc[:,(solver,'num_of_coefficients_total')] = df[(solver,'mcsat-explanation-onecellcad_num_of_coefficients_total')].fillna(0).astype(int)
    df.loc[:,(solver,'num_of_resultants')] = df[(solver,'mcsat-explanation-onecellcad_num_of_resultants')].fillna(0).astype(int)
    df.loc[:,(solver,'num_of_resultants_total')] = df[(solver,'mcsat-explanation-onecellcad_num_of_resultants_total')].fillna(0).astype(int)
    df.loc[:,(solver,'num_of_discriminants')] = df[(solver,'mcsat-explanation-onecellcad_num_of_discriminants')].fillna(0).astype(int)
    df.loc[:,(solver,'num_of_discriminants_total')] = df[(solver,'mcsat-explanation-onecellcad_num_of_discriminants_total')].fillna(0).astype(int)

    df.loc[:,(solver,'num_of_levels')] = df[(solver,'mcsat-explanation-onecellcad_num_of_levels')].fillna(0).astype(int)
    
df = df.join(ev.virtual_best(df, ["MCSATOCLWH11","MCSATOCLWH12","MCSATOCLWH13"], "VB-LW", ['nl_called', 'oc_called', 'resultant_barriers', 'max_degree','max_degree_total','num_of_coefficients','num_of_coefficients_total','num_of_resultants','num_of_resultants_total','num_of_discriminants','num_of_discriminants_total', 'num_of_levels']))


df
```

```python
df1 = ev.filter_solved(ev.filter_solved(ev.filter_solved(df,"MCSATOCLWH11"),"MCSATOCLWH12"),"MCSATOCLWH13")

df1 = df1[(df1[("MCSATOCLWH11","oc_called")]>0) & (df1[("MCSATOCLWH12","oc_called")]>0) & (df1[("MCSATOCLWH13","oc_called")]>0)]

df1 = df1[(df1[("MCSATOCLWH11","nl_called")]==0) & (df1[("MCSATOCLWH12","nl_called")]==0) & (df1[("MCSATOCLWH13","nl_called")]==0)]

len(df1)
```

```python
def get_or_zero(data,key):
    if key in data:
        return data[key]
    else:
        return 0

solvers = ["MCSATOCLWH11","MCSATOCLWH12","MCSATOCLWH13", "VB-LW"]
labels = ["LW-EQ-BC","LW-EQ-CH","LW-EQ-LDB","VB-LW"]
#cols = ['nl_called', 'oc_called', 'resultant_barriers', 'max_degree','max_degree_total','num_of_coefficients','num_of_coefficients_total','num_of_resultants','num_of_resultants_total','num_of_discriminants','num_of_discriminants_total']
cols = ['mean number of cells per instance', 'mean dimension of constructed cells', 'mean max degree per instance','mean number of resultants per instance','mean number of discriminants per instance','mean number of coefficients per instance']
data = []


for s in solvers:
    tmp=[]
    tmp.append(round(df1[(s,'oc_called')].mean(),2))
    tmp.append(round(df1[(s,'num_of_levels')].sum()/df1[(s,'oc_called')].sum(),2))
    tmp.append(round(df1[(s,'max_degree')].mean(),2))
    tmp.append(round(df1[(s,'num_of_resultants')].mean(),2))
    tmp.append(round(df1[(s,'num_of_discriminants')].mean(),2))
    tmp.append(round(df1[(s,'num_of_coefficients')].mean(),2))

    data.append(tuple(tmp))

summary = pd.DataFrame(data, index=labels,columns=cols)

summary
```
